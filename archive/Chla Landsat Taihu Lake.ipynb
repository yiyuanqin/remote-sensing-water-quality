```python
def getBandInfo(coordinates, start_date, end_date, band_selection, assetID = "LANDSAT/LE07/C01/T1_SR"):
    '''
    assetID: personal or GEE asset, default is USGS Landsat 8 Surface Reflectance Tier 1: "LANDSAT/LC08/C01/T1_SR"
    start_date, end_date: timestamp object
    band selection: (for bands in Landsat 8 with 30 m resolution)
        4: b1 blue
        5: b2 green
        6: b3 red
        7: b4 NIR
    '''
    try: 
        image = ee.ImageCollection(assetID).filterDate(start_date, end_date)
        point = ee.Geometry.Point(coordinates)
        get_point = image.getRegion(point,1)
        return get_point.get(1).getInfo()[band_selection]
        
    except: 
        return None
```

```python
getBandInfo(taihu.coordinates[0], taihu.image_start_date[0], taihu.image_end_date[0], 5, assetID = "LANDSAT/LE07/C01/T1_SR")
```

```python
# New column with b1 values from landsat 7
taihu['ls7_b1'] = taihu.apply(lambda x: getBandInfo(x['coordinates'], x['image_start_date'], x['image_end_date'], 4), axis=1)
```

```python
# New column with b2 values from landsat 7
taihu['ls7_b2'] = taihu.apply(lambda x: getBandInfo(x['coordinates'], x['image_start_date'], x['image_end_date'], 5), axis=1)
```

```python
# New column with b2 values from landsat 7
taihu['ls7_b3'] = taihu.apply(lambda x: getBandInfo(x['coordinates'], x['image_start_date'], x['image_end_date'], 6), axis=1)
```
