```python
def getBandInfo(coordinates, start_date, end_date, band_selection, assetID = "LANDSAT/LC08/C01/T1_SR"):
    '''
    assetID: personal or GEE asset, default is USGS Landsat 8 Surface Reflectance Tier 1: "LANDSAT/LC08/C01/T1_SR"
    start_date, end_date: timestamp object
    band selection: (for bands in Landsat 8 with 30 m resolution)
        4: b1
        5: b2
        6: b3
        7: b4
        8: b5
        9: b6
        10: b7
    '''
    try: 
        image = ee.ImageCollection(assetID).filterDate(start_date, end_date)
        point = ee.Geometry.Point(coordinates)
        get_point = image.getRegion(point,1)
        return get_point.get(1).getInfo()[band_selection]
    except: 
        return None
```

```python
# New column with b1 values from landsat 8
wi_lakes_select['ls8_b1'] = wi_lakes_select.apply(lambda x: getBandInfo(x['coordinates'], x['image_start_date'], x['image_end_date'], 4), axis=1)
```

```python
# New column with b5 values from landsat 8
wi_lakes_select['ls8_b5'] = wi_lakes_select.apply(lambda x: getBandInfo(x['coordinates'], x['image_start_date'], x['image_end_date'], 8), axis=1)
```

Clean up the data frame to contain only those with valid reflectance values and remove irrelevant columns

```python
# Reflectance values should be greater than 0. Negative values are computational artifacts. 
# https://landsat.usgs.gov/why-negative-values-over-water-landsat-surface-reflectance
clean_reflectance = (wi_lakes_select.ls8_b1 > 0) & (wi_lakes_select.ls8_b2 > 0)  & (wi_lakes_select.ls8_b3 > 0)  & (wi_lakes_select.ls8_b4 > 0) & (wi_lakes_select.ls8_b5 > 0) 
```

```python
# Drop irrelevant columns and select only valid reflectance values
wi_lakes_ls8_clean = wi_lakes_select[clean_reflectance].drop(columns=['index','group_seq_no','secchi__feet', 'secchi_hit_bottom', 'secchi__meters', 'total_phosphorus_ug_l','secchi_tsi','total_phosphorus_tsi'])
```
